```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
#read the data
df2 =  pd.read_csv("ht.csv")
#insert corresponding file
```

```python
df2.head()
```

```python
#Here National Typical Household, which assumes a household income of $61,828 (the national median household income), a national average household size of 2.62 and a national average number of commuters per household of 1.22.
#So we are assuming 2.62 as average size in order to account for individual cost but this assumption is only for code and national typical. For the data, I have done manual calculation and mapping to provide accurate information on value. The value for manual mapping is from regional typical which is available on the same folder named typicalhhsize. You can refer to that file if you need to perform the calculation.
df2['cost_per_individual'] = (df2['t_cost_ami'] /2.72).round(2)
```

```python
print(df2)
```

```python
#We can get the annual cost after dividing by average size.
#After that we are diving by 12 in order to account for monthly cost
df2['Monthly_cost'] = (df2['cost_per_individual']/12).round(0)
```

```python
print(df2)
```

```python
#Accounting for households with various adults ranging from HH! to HH7
df2["Transport_1"] = df2["Monthly_cost"]
df2["Transport_2"] = (df2["Transport_1"] * 2)
df2["Transport_3"] = (df2["Transport_1"] * 3)
df2["Transport_2"] = (df2["Transport_1"] * 4)
df2["Transport_5"] = (df2["Transport_1"] * 5)
df2["Transport_6"] = (df2["Transport_1"] * 6)
df2["Transport_7"] = (df2["Transport_1"] * 7)
```

```python
print(df2)
del df2['t_cost_ami']
del df2['cost_per_individual']
del df2['Monthly_cost']
```

```python
print(df2)
```
